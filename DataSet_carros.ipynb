```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('CarsData.csv')


print(df)


df_average_price = df.groupby('year')['price'].mean()

plt.figure(figsize=(10, 6))
plt.bar(df_average_price.index, df_average_price.values)
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Price of Cars by Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

data_cars = pd.read_csv('CarsData.csv')

data_cars['year'] = pd.to_datetime(data_cars['year'], format='%Y')

mean_price_by_decade = data_cars.groupby(data_cars['year'].dt.year // 10 * 10)['price'].mean().reset_index()

mean_price_by_decade = mean_price_by_decade.rename(columns={'price': 'average_salary'})

plt.figure(figsize=(10, 6))
plt.plot(mean_price_by_decade['year'], mean_price_by_decade['average_salary'], marker='o', linestyle='-')
plt.title('Média Salarial em base nos valores dos carros por suas década')
plt.xlabel('Década')
plt.ylabel('Média Salarial')
plt.grid(True)
plt.xticks(mean_price_by_decade['year'])
plt.tight_layout()
plt.show()
```
